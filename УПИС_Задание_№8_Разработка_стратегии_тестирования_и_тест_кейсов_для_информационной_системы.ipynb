# СТРАТЕГИЯ ТЕСТИРОВАНИЯ И ТЕСТ-КЕЙСЫ  
**Информационный портал журнала "Геодезия и аэрофотосъёмка"**

## 1. СТРАТЕГИЯ ТЕСТИРОВАНИЯ

### 1.1. Цели и задачи
Целью тестирования является обеспечение соответствия информационного портала журнала "Геодезия и аэрофотосъёмка" функциональным и нефункциональным требованиям, определенным в спецификации требований (SRS), а также выявление критических дефектов до развертывания системы на продуктивном сервере МИИГАиК.

Основные задачи:
- Проверка корректности работы аутентификации через MiigaikSSO
- Верификация функциональности управления выпусками и статьями
- Тестирование качества полнотекстового поиска по архиву
- Проверка корректности обработки и индексации PDF-файлов
- Оценка производительности системы при высокой нагрузке
- Проверка безопасности системы от несанкционированного доступа
- Оценка удобства использования интерфейса для целевой аудитории

### 1.2. Область тестирования

**Включено в тестирование:**
- Функциональность аутентификации через MiigaikSSO для редакторов и администраторов
- Управление выпусками (создание, публикация, архивация)
- Работа со статьями (добавление, редактирование, индексация PDF)
- Полнотекстовый поиск и фильтрация статей
- Просмотр и скачивание статей
- Экспорт метаданных статей
- Производительность системы при различных нагрузках
- Безопасность системы (защита от XSS, CSRF, SQL-инъекций)
- Кроссбраузерная совместимость
- Адаптивность интерфейса для разных устройств
- Надежность системы (восстановление после сбоев)

**Исключено из тестирования:**
- Функциональность самой системы MiigaikSSO (считается протестированной поставщиком)
- Сторонние библиотеки и фреймворки (Django, PostgreSQL, Redis)
- Физическая инфраструктура серверов МИИГАиК
- Долгосрочная стабильность системы (тестирование на месяцы/годы)

### 1.3. Виды тестирования

**Функциональное тестирование**
Метод: Комбинация ручного и автоматизированного тестирования
Охват: Все требования с приоритетом "Критический" и "Высокий"
Ответственный: Тестировщик-разработчик

**Интеграционное тестирование**
Подход: "Снизу вверх" (Bottom-up)
Фокус: Интеграция с MiigaikSSO, работа с PDF-файлами, взаимодействие между компонентами
Инструменты: pytest, Postman

**Тестирование производительности**
Виды: Нагрузочное, стрессовое
Инструменты: Apache JMeter, Locust
Критерии: Время отклика ≤ 2с при 100 одновременных пользователях

**Тестирование безопасности**
Виды: Проверка аутентификации, тестирование на уязвимости (OWASP Top 10)
Инструменты: OWASP ZAP, sqlmap
Ответственный: Разработчик с привлечением специалиста безопасности МИИГАиК

**Тестирование юзабилити**
Метод: Тестирование с привлечением реальных пользователей (сотрудники и студенты МИИГАиК)
Цель: Оценка удобства интерфейса для целевой аудитории
Критерии: Не менее 90% успешного выполнения основных сценариев

**Регрессионное тестирование**
Частота: После каждого исправления дефектов и добавления новой функциональности
Автоматизация: 60% регрессионных тестов автоматизировано
Инструменты: Selenium, pytest

### 1.4. Инструменты тестирования

| Тип тестирования | Инструмент | Обоснование |
|-----------------|------------|-------------|
| Управление тест-кейсами | TestRail (бесплатная академическая лицензия) | Централизованное хранение тест-кейсов, отслеживание прогресса |
| Функциональное автоматизированное | Selenium, pytest | Открытые инструменты, хорошая интеграция с Python/Django |

| Безопасность | OWASP ZAP, Burp Suite Community | Open-source инструменты для тестирования безопасности |
| Статический анализ кода | SonarQube | Выявление потенциальных дефектов и уязвимостей на этапе разработки |
| Мониторинг | Prometheus + Grafana | Отслеживание производительности в тестовой среде |

### 1.5. Критерии входа и выхода

**Entry Criteria (Критерии начала тестирования):**
- Код зафиксирован в системе контроля версий (Git)
- Сборка развернута на тестовом стенде
- Тестовая документация (тест-кейсы) подготовлена и утверждена руководителем
- Тестовые данные созданы и доступны
- Выполнены unit-тесты с покрытием ≥ 70% кодовой базы
- Пройден smoke test базовой функциональности

**Exit Criteria (Критерии завершения тестирования):**
- Выполнено ≥ 95% запланированных тест-кейсов
- Покрытие требований с приоритетом "Критический" = 100%
- Покрытие требований с приоритетом "Высокий" ≥ 90%
- Pass Rate ≥ 98% для критических и высокоприоритетных тестов
- Все дефекты с серьезностью "Critical" исправлены
- Все дефекты с серьезностью "Major" исправлены или отложены с обоснованием
- Нагрузочные тесты пройдены успешно (время отклика ≤ 2с при 100 пользователях)
- Тесты безопасности не выявили критических уязвимостей
- Юзабилити-тестирование показало успешное выполнение ≥ 90% сценариев

### 1.6. Метрики качества

**Метрики покрытия:**
- Покрытие требований: ≥ 90% для критических и высокоприоритетных требований
- Покрытие кода unit-тестами: ≥ 70%
- Покрытие функциональных сценариев: ≥ 85%

**Метрики дефектов:**
- Плотность дефектов: < 2 дефекта/1000 строк кода
- Эффективность обнаружения дефектов (DDE): ≥ 90%
- Распределение дефектов по серьезности:
  - Critical: 0 (должны быть исправлены)
  - Major: ≤ 5%
  - Minor и Trivial: оставшиеся

**Метрики выполнения:**
- Процент выполненных тестов: ≥ 95%
- Pass Rate: ≥ 98% для критических тестов
- Скорость исправления дефектов:
  - Critical: < 24 часа
  - Major: < 72 часа
  - Minor: < 1 неделя

## 6. КРИТЕРИИ ПРОХОЖДЕНИЯ ТЕСТИРОВАНИЯ

Система считается готовой к релизу при выполнении всех следующих условий:

   - Все требования с приоритетом "Критический" покрыты тест-кейсами и успешно пройдены
   - Не менее 90% требований с приоритетом "Высокий" покрыты и успешно пройдены

   - Время индексации PDF-файла размером до 15 МБ не превышает 30 секунд
   - Система выдерживает нагрузку до 100 одновременных пользователей без деградации производительности

   - Система успешно прошла проверку OWASP ZAP без критических уязвимостей
   - Аутентификация через MiigaikSSO функционирует корректно, все сессии защищены
   - Все данные передаются по защищенному протоколу HTTPS
   - Личные данные пользователей хранятся в зашифрованном виде

   - Все функции доступны для пользователей с ограниченными возможностями (соответствие WCAG 2.1 AA)

5. **Документация:**
